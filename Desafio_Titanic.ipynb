Você deve ter ouvido falar do famoso navio de luxo RMS Titanic, que afundou tristemente depois de atingir um iceberg em 15 de abril de 1912, em sua primeira viagem.

<b>Vamos explorar alguns fatos impressionantes sobre o Titanic</b>

Em 30 de abril de 1907, em um jantar entre J Bruce Ismay e Lord William James Pirrie, surgiu a idéia de construir o Titanic.

<b>Titânico</b>

Eles queriam construir um navio ainda melhor do que a linha de luxo Cunard no Atlântico.

<b>Fatos surpreendentes sobre o Titanic</b>

Depois que eles decidiram o que iriam fazer, o planejamento foi a todo vapor.

- Foram necessários três anos para construir e custar 7.500.000 dolares (o  equivalente a 400 milhões de dolares hoje).

- 8 dos 3000 trabalhadores da construção civil - que recebiam US $ 10 por semana - morreram durante a construção.

O Titanic foi construído em Belfast no estaleiro Harland and Wolff pela empresa White Star Line. Era de propriedade do JP Morgan, um magnata americano.

<b>Titanic em Harland e Wolff</b>

Titanic em construção no estaleiro Harland and Wolff

<b> O grande lançamento</b>

Após o seu lançamento em Belfast, Irlanda, em 31 de maio de 1911, na presença de cerca de 90 membros da tripulação e uma multidão de cerca de 100.000 pessoas, ela levou 62 segundos para percorrer a pista deslizante sobre 22 toneladas de sabão e sebo.

Foi uma grande ocasião. Em 2 de abril de 1912, o Titanic partiu de Belfast para Southampton.

Em 10 de abril de 1912, partiu das docas de Southampton em sua primeira travessia do Atlântico para Nova York.

A viagem durou 137 horas.

O capitão era o capitão Edward J. Smith, que era um mestre no mar com mais de 43 anos de experiência.

Muitas pessoas dizem que ele realmente ignorou os avisos de iceberg enquanto tentava estabelecer um novo recorde de tempo. Isso não é verdade. Isso seria muito irresponsável!

<b>Como era o Titanic? </b>

O Titanic tinha quase o comprimento de três campos de futebol, tão alto quanto um edifício de 17 andares, com quatro chaminés; apenas uma das chaminés era para decoração para dar ao navio uma aparência equilibrada.

Os motores usavam mais de 800 toneladas de carvão por dia e a velocidade máxima era de 24 nós (27 milhas por hora).

Ela foi uma das primeiras naves a ter sistema telefônico e luzes elétricas em todos os quartos, com quatro elevadores, piscina aquecida, academia, duas bibliotecas e duas barbearias.

Havia 840 cabines, 9 decks e 10.000 lâmpadas.

<b>Custos com ingressos</b>

O custo de um bilhete de primeira classe foi de 4.350 dolares (875 libras). Era de 1.750 dolares para um bilhete de classe padrão e 30 dolares para um bilhete de terceira classe.

Os ingressos da Terceira Classe eram principalmente para famílias que estavam emigrando para os Estados Unidos.

Em muitos casos, eles venderam tudo para entrar no Titanic e dar-lhes algum dinheiro para começar na América.

Aqueles eram alguns bilhetes seriamente caros!


<b>Tudo sobre passageiros e tripulação</b>

O Titanic poderia transportar cerca de 3.547 pessoas quando totalmente carregado, que são passageiros e tripulação, mas eles partiram de Southampton com 2.223 a bordo.

- 1.324 eram passageiros e 885 eram tripulantes.

- Havia 13 casais em lua de mel também.

Quando o Titanic afundou, o número total de sobreviventes era de cerca de 705. Destes, 492 eram passageiros, a maioria deles da Primeira Classe, juntamente com 214 tripulantes.

Todos os 25 oficiais de engenharia morreram enquanto tentavam manter o navio à tona.

Dois cães sobreviveram e a banda tocou até o navio afundar.

Infelizmente, nenhum dos membros da banda sobreviveu.

<b>Que suprimentos o Titanic tinha a bordo?</b>

Aqui estão apenas alguns dos suprimentos que o Titanic carregava (ela certamente carregava muito!):

- 75.000 libras de carne fresca

- 36.000 maçãs

- 40 toneladas de batatas

- 1.000 pães

- 10.000 libras de açúcar

- 40.000 ovos frescos

- 6.000 libras de manteiga

- 15.000 garrafas de cerveja

<b>O naufrágio do Titanic</b>

Em 14 de abril de 1912, o Titanic recebeu avisos de icebergs na área de outros navios. 

Após o segundo aviso do iceberg, o navio mudou de rumo, o que acabou sendo um grande erro, uma vez que o levou a entrar em rota de colisão com o iceberg.

Mais avisos vieram, mas o capitão Smith não os recebeu. Ele também não recebeu o último às 21h40 da noite.

Às 23h, a tripulação do navio "californiano" enviou uma mensagem dizendo que havia parado de navegar devido a icebergs, mas o capitão em exercício não estava interessado.

O iceberg fatal estava a 24 quilômetros de distância. Não havia vento e as águas estavam calmas. Na verdade, isso tornou mais difícil detectar icebergs, já que não era possível ver água quebrando contra ele.

O Titanic atingiu o iceberg às 23h39.

## Valores faltantes

```python
# Percentual de valores ausentes das variáveis do nosso conjunto de dados

round((df_train.isnull().sum() / df_train.shape[0]).sort_values(ascending=False) * 100,2)
```

## Preencher os valores nulos

```python
# Atribuindo valores médios para os registros nulos das colunas AGE, FARE.
# Para a coluna EMBARKED atribuido a ocorrencia mais comum.

# age
age_median = df_train['age'].median()
df_train['age'].fillna(age_median, inplace=True)

# fare
fare_median = df_train['fare'].median()
df_train['fare'].fillna(fare_median, inplace=True)

# embarked
embarked_top = df_train['embarked'].value_counts()[0]
df_train['embarked'].fillna(embarked_top, inplace=True)
```

```python
# dummie variables para 'Embaked'
embarked_dummies = pd.get_dummies(df_train['embarked'], prefix='embarked')
df_train = pd.concat([df_train, embarked_dummies], axis=1)
df_train.drop('embarked', axis=1, inplace=True)

display(df_train)
```

```python
# Removendo a coluna EMBARKED_644
df_train.drop('embarked_644', axis=1, inplace=True)
```

```python
# Função para Converter idade menor que 1 para zero

def age_lt_zero(age):
  if age < 1:
    return 0
  else:
    return age


df_train['age'] = df_train['age'].apply(age_lt_zero) 

df_train['age']
```

```python
# Verificando a MÉDIA, o MÁXIMO, o MÍNIMO e a MEDIANA da Idade por Sexo
df_train.groupby(['sex'])['age'].agg(['max','min','mean','median','count'])
```

```python
df_train.isnull().sum().sort_values(ascending=False)
```

```python
# converter 'Sex' em 0 e 1
df_train['sex'] = df_train['sex'].map({'male': 0, 'female': 1})
```

# Criando uma coluna para faixa etária

```python
def faixa_etaria(c):
  if c['age'] < 20:
    return 1
  else:
     if c['age'] < 40:
       return 2
     else:    
       if c['age'] < 60:
         return 3
       else:
         return 4


df_train['faixa_etaria'] = df_train.apply(faixa_etaria, axis=1)
df_train.head()
```

```python
# Verificando as quantidades por faixa etária
df_train['faixa_etaria'].value_counts()
```

```python
df_train['age'].value_counts().sort_values(ascending=False)
```

```python
# Verificando as quantidades de sobreviventes por faixa etaria e sexo 
df_train.groupby(['sex','survived'])['faixa_etaria'].agg('value_counts')
```

## Excluindo colunas que não serão utilizadas

```python
# Excluindo colunas que não são úteis

# Grande quantidade de valores nulos
df_train.drop('cabin', axis=1, inplace=True)

# Nome não é considerado para avaliação
df_train.drop('name', axis=1, inplace=True)

# Id Kaggle e Número do Ticket não intereferem no resultado
df_train.drop(['Kaggle_Pass_Id',  'ticket'], axis=1, inplace=True)

# Nr do Barco Salva Vida, Nr do corpo, Casa/Detino, Sequencial 
df_train.drop(['boat', 'body', 'home.dest', 'orig_seq'], axis=1, inplace=True)
```
